#### Merging the datasets for better cross-section analysis

```python
# Merging function
def merge_data(tickers):
    # Load the data for each ticker
    data_frames = {ticker: load_data("2-year sample/"f'{ticker}_merged.csv') for ticker in tickers}

    # Rename the columns in each DataFrame except 'date' and 'minute'
    for ticker, df in data_frames.items():
        df.rename(columns=lambda name: f"{ticker}_{name}" if name not in ['date', 'minute'] else name, inplace=True)

    # Initialize merged DataFrame as the first ticker's DataFrame
    merged_df = list(data_frames.values())[0]

    # Merge each DataFrame on 'date' and 'minute'
    for ticker, df in data_frames.items():
        if ticker == list(data_frames.keys())[0]:  # Skip the first ticker as it is already in merged_df
            continue
        merged_df = pd.merge(merged_df, df, on=['date', 'minute'], how='outer')

    return merged_df

# List of tickers
tickers = ['aapl', 'amd', 'amzn', 'fb', 'msft', 'nflx', 'nvda', 'tsla', 'v', 'xom']

# Call the merge function
combined_df = merge_data(tickers)
combined_df.dropna(inplace=True)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Extract volume columns
volume_columns = [col for col in combined_df.columns if 'volume' in col]
volumes = combined_df[volume_columns]

# Standardize the volume data
scaler = StandardScaler()
volumes_standardized = scaler.fit_transform(volumes)

# Perform PCA
pca = PCA()
principal_components = pca.fit_transform(volumes_standardized)

# Calculate explained variance and cumulative explained variance
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance.cumsum()

# Find the number of components for optimal variance explained (>95% cumulative variance)
optimal_num_components = len(cumulative_explained_variance[cumulative_explained_variance < 0.95]) + 1

# Plotting the explained variance
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, where='mid', label='Cumulative explained variance')
plt.axvline(optimal_num_components, color='k', linestyle='--', label='Optimal number of components')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.title('Explained Variance by PCA Component')
plt.show()

# Output the principal components and the optimal number of components
print("Principal Components:\n", principal_components)
print("Optimal number of components:", optimal_num_components)
```

#### Construct minutely return series

```python
def calculate_adjacent_returns(df, ticker):
    # Calculate the simple returns
    df[f'{ticker}_return'] = df[f'{ticker}_close'].pct_change()

    # Find out if the current minute is adjacent to the previous one
    def is_adjacent(curr, prev):
        # If it's the start of an hour, check if previous is end of previous hour
        if curr % 100 == 0 and prev % 100 == 59 and (curr // 100 - prev // 100) == 1:
            return True
        # Else, check if it's just the next minute
        return curr - prev == 1

    # Determine if the current row is adjacent to the previous row
    df['is_adjacent'] = [is_adjacent(curr, prev) for curr, prev in zip(df['minute'], df['minute'].shift(fill_value=df['minute'].iloc[0] - 1))]

    # Set returns to NaN where minutes are not adjacent
    df.loc[~df['is_adjacent'], f'{ticker}_return'] = None

    # Drop the 'is_adjacent' column as it is no longer needed
    df.drop(columns='is_adjacent', inplace=True)
    return df[f'{ticker}_return']

# Reinitialize returns_df to just contain 'date' and 'minute'
returns_df = combined_df[['date', 'minute']].copy()

# Calculate returns for each ticker and add to the returns DataFrame
for ticker in tickers:
    returns_series = calculate_adjacent_returns(combined_df.copy(), ticker)
    returns_df[f'{ticker}_return'] = returns_series

# Drop rows where all returns are NaN across all tickers
returns_df.dropna(how='all', subset=[f'{ticker}_return' for ticker in tickers], inplace=True)

# Reset index to clean up the DataFrame
returns_df.reset_index(drop=True, inplace=True)
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

def run_arima(df, stock_name, order=(5,3,0), seasonal_order=(0,0,0,0), future_periods=10):
    df['minute_str'] = df['minute'].apply(lambda x: f"{int(x):04d}")
    
    # Create a datetime index from the date / minute to a more consumable format by ml
    df['datetime'] = pd.to_datetime(df['date'].astype(str) + ' ' + df['minute_str'], format='%Y-%m-%d %H%M')
    df.set_index('datetime', inplace=True)
    df.sort_index(inplace=True)
    df['date_only'] = df.index.date
    date_counts = df['date_only'].value_counts()
    
    plt.figure(figsize=(10, 5))
    date_counts.plot(kind='bar')
    plt.title('Data Distribution Across Dates')
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.show()

    close_price = df[f'{stock_name}_close']
    model = SARIMAX(close_price, order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
    results = model.fit(disp=False)
    
    last_date = df.index[-1]
    future_dates = pd.date_range(start=last_date, periods=future_periods + 1, freq='T')[1:]
    forecast = results.get_forecast(steps=future_periods)
    mean_forecast = forecast.predicted_mean
    conf_int = forecast.conf_int()
    
    mean_forecast.index = future_dates
    conf_int.index = future_dates
    
    plt.figure(figsize=(10, 5))
    plt.plot(close_price.tail(100).index, close_price.tail(100), label='Observed')
    plt.plot(mean_forecast.index, mean_forecast, color='r', label='Forecast')
    plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='pink', label='Confidence Interval')
    plt.title(f'Forecasting the next {future_periods} periods for {stock_name}')
    plt.legend()
    plt.show()

    return results.summary()

# Example Usage
if 'combined_df' in locals():
    result_summary = run_arima(combined_df, 'aapl') 
    print(result_summary)
else:
    print("DataFrame not found. Please ensure your DataFrame is loaded and named correctly.")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt

def run_lstm_and_plot(df, stock_name, lstm_units=50, dense_units=25, batch_size=32, epochs=20, seq_length=10, future_periods=30):
    df['date'] = df['date'].astype(str)
    df['minute'] = df['minute'].apply(lambda x: f"{int(x):04d}")

    # Create a datetime index from the date / minute to a more consumable format by ml
    df.set_index(pd.to_datetime(df['date'] + ' ' + df['minute'], format='%Y-%m-%d %H%M'), inplace=True)
    df.drop(['date', 'minute'], axis=1, inplace=True)

    # Normalize the data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df)

    # Function to create sequences
    def create_sequences(data, seq_length):
        xs = []
        ys = []
        for i in range(len(data)-seq_length-1):
            x = data[i:(i+seq_length)]
            y = data[i+seq_length]
            xs.append(x)
            ys.append(y)
        return np.array(xs), np.array(ys)

    # Create sequences
    X, y = create_sequences(scaled_data, seq_length)

    # Split the data
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(lstm_units, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(LSTM(lstm_units))
    model.add(Dense(dense_units))
    model.add(Dense(y_train.shape[1]))

    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test))

    # Making predictions
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(predictions)  # Inverse transform 

    y_test_actual = scaler.inverse_transform(y_test)

    # Predict future periods
    last_sequence = scaled_data[-seq_length:]
    current_sequence = last_sequence.reshape((1, seq_length, X_train.shape[2]))
    future_predictions = []

    for _ in range(future_periods):
        next_prediction = model.predict(current_sequence)[0]
        future_predictions.append(next_prediction)
        # Reshape next_prediction to match the 3D shape expected by LSTM
        next_prediction_reshaped = next_prediction.reshape((1, 1, X_train.shape[2]))
        current_sequence = np.append(current_sequence[:, 1:, :], next_prediction_reshaped, axis=1)

    future_predictions = scaler.inverse_transform(future_predictions)


    # Plotting the results
    plt.figure(figsize=(12, 6))
    total_sequence = np.concatenate((y_test_actual[:, 0], future_predictions[:, 0]))
    plt.plot(np.arange(len(y_test_actual)), y_test_actual[:, 0], label='Actual Close Price')
    plt.plot(np.arange(len(y_test_actual), len(y_test_actual) + future_periods), future_predictions[:, 0], label='Future Predictions', color='red')
    plt.title(f'{stock_name} Stock Price Prediction and Future Forecast')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

    return model

# Example usage
model_aapl = run_lstm_and_plot(aapl_df, 'AAPL', lstm_units=64, dense_units=32, batch_size=16, epochs=5, seq_length=20, future_periods=30)
model_amd = run_lstm_and_plot(amd_df, 'AMD', lstm_units=32, dense_units=16, batch_size=16, epochs=5, seq_length=15, future_periods=30)
```

```python
def run_random_forest(df, stock_name, n_estimators=100, max_depth=None, random_state=42, future_periods=30):
    # Ensure date and minute are properly formatted as strings
    df['datetime'] = pd.to_datetime(df['date'].astype(str) + ' ' + df['minute'].astype(str).str.zfill(4), format='%Y-%m-%d %H%M')
    df.set_index('datetime', inplace=True)
    df.drop(['date', 'minute'], axis=1, inplace=True)

    # Split the data into features and target
    X = df.drop(f'{stock_name}_close', axis=1)
    y = df[f'{stock_name}_close']
    
    # Split the dataset into training and testing sets
    split_point = int(len(X) * 0.8)
    X_train, X_test = X[:split_point], X[split_point:]
    y_train, y_test = y[:split_point], y[split_point:]

    # Initialize and train the Random Forest model
    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, random_state=random_state)
    model.fit(X_train, y_train)

    # Predictions
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    
    # Future prediction setup
    future_X = pd.DataFrame(np.tile(X_test.iloc[-1], (future_periods, 1)), columns=X_test.columns)
    future_predictions = model.predict(future_X)

    # Plotting the results
    plt.figure(figsize=(12, 6))
    total_sequence = np.concatenate((y_test.to_numpy(), future_predictions))
    plt.plot(np.arange(len(y_test)), y_test, label='Actual Close Price')
    plt.plot(np.arange(len(y_test), len(y_test) + future_periods), future_predictions, label='Future Predictions', color='green')
    plt.title(f'{stock_name} Stock Price Prediction and Future Forecast using Random Forest')
    plt.xlabel('Time')
    plt.ylabel('Close Price')
    plt.legend()
    plt.show()

    print(f'MSE for {stock_name}: {mse}')
    return model

model_aapl = run_random_forest(aapl_df, 'aapl', n_estimators=100, max_depth=10, future_periods=30)
```
